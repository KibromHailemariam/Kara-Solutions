# Introduction to Credit Risk
   - Credit risk is the risk of loss due to a borrower's failure to repay a loan or meet contractual obligations. This section delves into how credit risk is managed in the banking sector, particularly for Bati Bank's buy-now-pay-later initiative, managing credit risk is crucial to ensure sustainability and minimize defaults.

#### Key Challenges in BNPL Credit Risk Management:

*	- Limited Credit Histories: Many BNPL users may lack traditional credit records.
*	- Short-Term Credit Cycles: Quick repayment terms require faster risk assessment.
*	- High Transaction Volume: Numerous small-value transactions complicate risk monitoring.

#### Bati Bankâ€™s Risk Mitigation Strategies:

*	- AI-Based Credit Scoring: Leveraging alternative data (e.g., payment history) for borrower assessment.
*	- Dynamic Credit Limits: Adjusting limits based on repayment behavior.
*	- Merchant Collaboration: Sharing risk to maintain operational stability.
*	- Default Recovery Mechanisms: Automated reminders, flexible payment plans, and recovery processes.


### Key Concepts:

### Credit Scoring 
 - Definition: A method to evaluate the creditworthiness of potential borrowers.
 - Purpose: Predicts the likelihood of default based on historical data.


### Default
   - Definition: Failure to pay the loan principal or interest on time.
   - Variation: Definitions can differ but must adhere to regulatory standards.


### Basel II Capital Accord
   - Overview: Regulatory framework for banking supervision, focusing on capital adequacy.
   - Pillars: 
       - Pillar 1 - Minimum Capital Requirements
       - Pillar 2 - Supervisory Review Process
       - Pillar 3 - Market Discipline


## Visualization

### imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Code Examples:
  - we'll simulate data based on the given column definitions:

```python

# Simulate data based on column definitions
np.random.seed(42)
n_samples = 1000

data = {
    'TransactionId': range(1, n_samples + 1),
    'BatchId': np.random.randint(1, 500, n_samples),
    'AccountId': np.random.randint(1, 100, n_samples),
    'SubscriptionId': np.random.randint(1, 50, n_samples),
    'CustomerId': np.random.randint(1, 100, n_samples),
    'CurrencyCode': np.random.choice(['USD', 'EUR', 'GBP'], n_samples),
    'CountryCode': np.random.randint(1, 250, n_samples),
    'ProviderId': np.random.randint(1, 20, n_samples),
    'ProductId': [f'Product_{i}' for i in np.random.randint(1, 50, n_samples)],
    'ProductCategory': np.random.choice(['Electronics', 'Clothing', 'Books'], n_samples),
    'ChannelId': np.random.choice(['web', 'Android', 'IOS', 'pay later', 'checkout'], n_samples),
    'Amount': np.random.uniform(-1000, 1000, n_samples).round(2),
    'Value': np.abs(np.random.uniform(0, 1000, n_samples)).round(2),
    'TransactionStartTime': np.datetime64('2023-01-01') + np.random.randint(0, 365, n_samples).astype('timedelta64[D]'),  # Changed here
    'PricingStrategy': np.random.choice(['Standard', 'Premium'], n_samples),
    'FraudResult': np.random.choice([0, 1], n_samples, p=[0.95, 0.05])  # Assuming 5% fraud rate
}

df = pd.DataFrame(data)

# Display first few rows of the dataset
print(df.head())
```

## Visualize Examples:

```python

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Example 1: Distribution of Transaction Amounts to see skewness
sns.histplot(df['Amount'], kde=True, bins=50)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Example 2: Correlation between Amount and Fraud (simplified for illustration)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Amount', y='FraudResult', alpha=0.1)
plt.title('Amount vs. Fraud Occurrence')
plt.xlabel('Amount')
plt.ylabel('Fraud (1 = Yes, 0 = No)')
plt.show()

# Example 3: Distribution of transactions by Product Category
plt.figure(figsize=(12, 6))
df['ProductCategory'].value_counts().plot(kind='bar')
plt.title('Transactions by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Transactions')
plt.show()
```

### These visualizations:
* Distribution of Transaction Amounts: This can help understand the skewness in transaction data, which might be relevant for understanding credit behavior.

* Amount vs. Fraud: A scatter plot to see if there's any visible pattern between transaction amount and fraud, although in reality, this relationship would be more nuanced.

* Transactions by Product Category: Shows which product categories are more commonly purchased, potentially indicating customer preferences or spending habits that might correlate with credit risk.
